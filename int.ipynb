```python
#WAP TO CHECK WHETHER GIVEN NUMBER IS A SPY NUMBER OR NOT ?  
a = int(input("enter a number : "))
sum =  0
product = 1
while a > 0:
    dig = a % 10
    sum = sum + dig
    product = product * dig
    a = a // 10
print("enter the sum : ",sum)
print("enter the product : ",product)
if product == sum :
    print("it is a spy number")
else:
    print("it is not a spy number")
```

```python
#WAP TO CHECK WHETHER GIVEN NUMBER IS A PERFECT NUMBER OR NOT
n = int(input("enter a number : "))
sum = 0
for i in range(1, n):
    if n % i == 0:
        sum += i
print("enter a number : ",sum)
if sum == n:
    print("it is a perfect number.")
else:
    print("it is a not perfect number.")
```

```python
#WAP TO CHECK WHETHER GIVEN NUMBER IS A PERFECT FACTORIAL NUMBER OR NOT
n = int(input("enter a number : "))
m = 1
for i in range(1, n + 1):
    m = m * i
print("enter the number : ",n)
print("enter the factorial number : ",m)
```
